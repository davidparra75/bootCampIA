```python
#quiero ver cual departamento tiene mas usuarios

usuarios_departamento = df.groupby("Departamento")["Usuarios"].sum().sort_values(ascending=False)
usuarios_departamento

plt.Figure(figsize=(12,6))
usuarios_departamento.plot(kind='bar', color='red')
plt.title("usuarios por departamento")
plt.xlabel("Departamento")
plt.ylabel("Ciudad")
plt.show()
```

```python
#cantidad de wats producidas por departamento
watts_departamento = df.groupby("Departamento")["Energía [kWh/día]"].sum().sort_values(ascending=False)
watts_departamento
```

```python
#grafico de pie para vizualizar aporte de kw al dia por departamento

plt.Figure(figsize=(100,60))
watts_departamento.plot(kind="pie",labels=df["Departamento"], title="Diagrama de  Watts por departamento", autopct='%1.1f%%')
```
